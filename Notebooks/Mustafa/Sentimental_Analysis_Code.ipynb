```python
# Clean text
def clean_text(text):
    text = text.lower()
    text = re.sub(r'http\\S+', '', text)  # Remove URLs
    text = re.sub(r'[^a-zA-Z0-9.,!?$&@#\s]', '', text)  # Remove only unnecessary special characters
    text = re.sub(r'\s+', ' ', text).strip()  # Remove extra spaces
    return text

df_updated['comment_body'] = df_updated['comment_body'].apply(clean_text)
```

```python
# Initialize spell checker
spell = SpellChecker()

# Function to correct spelling in text with curly quote normalization
def correct_spelling_spellchecker(text):
    if not isinstance(text, str) or text.strip() == "":
        return text  # Return unchanged if it's not a valid string

    # Normalize curly quotes to straight quotes
    text = text.replace("‘", "'").replace("’", "'").replace("“", '"').replace("”", '"')

    # Tokenize the text
    tokens = word_tokenize(text)

    corrected_tokens = []
    for token in tokens:
        # Spell-check only purely alphabetic tokens
        if token.isalpha():
            corrected = spell.correction(token)
            corrected_tokens.append(corrected if corrected else token)
        else:
            corrected_tokens.append(token)

    return ' '.join(corrected_tokens)  # Rejoin tokens into a corrected sentence

# Apply spell checking and curly quote normalization to the 'comment_body' column
df_updated['comment_body'] = df_updated['comment_body'].apply(correct_spelling_spellchecker)

# Display an example output
print(df_updated.iloc[199]['comment_body'])
```

```python
# Setup
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
nlp = spacy.load("en_core_web_sm")

# Final preprocessing function: Tokenization, stopword removal, lemmatization, Part of Speech Tagging
def clean_comment(text):
    text = str(text)  # Convert NaN/float to empty string or stringified text

    doc = nlp(text)

    filtered_tokens = [
        token.lemma_.lower()
        for token in doc
        if not token.is_stop and token.is_alpha and token.pos_ in ['NOUN', 'VERB', 'ADJ']
    ]
    return ' '.join(filtered_tokens)
```

```python
df_updated['final_preprocessed_text'] = df_updated['comment_body'].apply(clean_comment)
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# Initialize VADER
analyzer = SentimentIntensityAnalyzer()

# Apply VADER sentiment on final preprocessed column
df_updated['vader_scores_model1'] = df_updated['final_preprocessed_text'].apply(analyzer.polarity_scores)

# Create a sentiment label column
def classify_sentiment(score):
    c = score['compound']
    if c >= 0.05:
        return 'positive'
    elif c <= -0.05:
        return 'negative'
    else:
        return 'neutral'

df_updated['sentiment_label_vader_model1'] = df_updated['vader_scores_model1'].apply(classify_sentiment)
```

```python
# Initialize VADER
sia = SentimentIntensityAnalyzer()

# Normalize the comments
df_updated['comment_text_cleaned'] = df_updated['comment_body'].astype(str).str.lower()

# Apply VADER sentiment analysis
df_updated['vader_sentiment_scores_model2'] = df_updated['comment_text_cleaned'].apply(sia.polarity_scores)

# Extract compound, negative, neutral, and positive scores into separate columns
df_updated['vader_compound_score_model2'] = df_updated['vader_sentiment_scores_model2'].apply(lambda x: x['compound'])
df_updated['vader_neg_score_model2'] = df_updated['vader_sentiment_scores_model2'].apply(lambda x: x['neg'])
df_updated['vader_neu_score_model2'] = df_updated['vader_sentiment_scores_model2'].apply(lambda x: x['neu'])
df_updated['vader_pos_score_model2'] = df_updated['vader_sentiment_scores_model2'].apply(lambda x: x['pos'])

# Function to classify sentiment
def classify_sentiment(compound):
    if compound >= 0.05:
        return 'Positive'
    elif compound <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'

# Apply classification
df_updated['vader_sentiment_class_model2'] = df_updated['vader_compound_score_model2'].apply(classify_sentiment)

# View results
print(df_updated[['comment_body',
                  'vader_compound_score_model2',
                  'vader_neg_score_model2',
                  'vader_neu_score_model2',
                  'vader_pos_score_model2',
                  'vader_sentiment_class_model2']].head())
```

```python
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch.nn.functional as F
import ast  # For safe string-to-list conversion

# Load tokenizer and model
tokenizer = AutoTokenizer.from_pretrained("cardiffnlp/twitter-roberta-base-sentiment")
model = AutoModelForSequenceClassification.from_pretrained("cardiffnlp/twitter-roberta-base-sentiment")
model.eval() # Set the model to evaluation mode

# use GPU if available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

# Label mapping
label_map = {0: "Negative", 1: "Neutral", 2: "Positive"}

# Prediction function
def predict_sentiment_from_tokens(input_ids_list):
    input_ids_tensor = torch.tensor(input_ids_list).unsqueeze(0).to(device)  # Add batch dimension
    with torch.no_grad(): # Disable gradient calculation for inference
        output = model(input_ids=input_ids_tensor)
        probs = F.softmax(output.logits, dim=1).squeeze().tolist() # Convert logits to probabilities
        label_idx = torch.argmax(output.logits, dim=1).item() # Get index of highest probability
        return label_map[label_idx], probs # Return predicted label and full probability list

# Convert tokens_ids column from string to list if necessary
df_updated["tokens_ids"] = df_updated["tokens_ids"].apply(ast.literal_eval)

# Apply prediction and unpack results
df_updated[["sentiment_twitter_roberta", "sentiment_probs_twitter_roberta"]] = df_updated["tokens_ids"].apply(
    lambda x: pd.Series(predict_sentiment_from_tokens(x))
)

# Store individual sentiment scores
df_updated["sentiment_score_twitter_roberta"] = df_updated["sentiment_probs_twitter_roberta"].apply(lambda x: max(x))
df_updated["negative_score_twitter_roberta"] = df_updated["sentiment_probs_twitter_roberta"].apply(lambda x: x[0])
df_updated["neutral_score_twitter_roberta"] = df_updated["sentiment_probs_twitter_roberta"].apply(lambda x: x[1])
df_updated["positive_score_twitter_roberta"] = df_updated["sentiment_probs_twitter_roberta"].apply(lambda x: x[2])
```

```python
# initial preprocessing
def initialPreprocess_dataframe(dataframe):
    df_no_duplicates = dataframe.drop_duplicates()
    df_no_duplicates.rename(columns={'productId': 'short_code'}, inplace=True)


    df_nlp = df_no_duplicates.copy()

    # drop duplicates
    df_nlp.rename(columns={'isVerified': 'isVerifiedPurchaser'}, inplace=True)

    # filtering based on verified purchaser
    df_nlp = df_nlp[df_nlp['isVerifiedPurchaser'] == True]

    #filtering out not useful feedbacks
    df_nlp = df_nlp[df_nlp['negativeFeedbackCount'] <= df_nlp['positiveFeedbackCount']]

    df_nlp = df_nlp.dropna(subset=['text'])  # Remove rows where 'text' is missing

    df_nlp['title'] = df_nlp['title'].fillna('')  # Fill missing 'title' with empty string

    df_updated = df_nlp

    # Group by productName and count likes based on rating >= 4
    # Calculate likes for each product
    likes_count = df_updated.groupby('productName')['rating'].apply(lambda x: (x >= 4).sum()).reset_index(name='likes')

    # Merge likes count back to the original DataFrame
    df_updated = df_updated.merge(likes_count, on='productName', how='left')

    # Split lastModificationTime into comment_year and comment_month
    df_updated['comment_year'] = pd.to_datetime(df_updated['lastModificationTime']).dt.year
    df_updated['comment_month'] = pd.to_datetime(df_updated['lastModificationTime']).dt.month

    # Add two new columns: category and source
    df_updated['category'] = 'Product'
    df_updated['source'] = 'Costco'


    df_updated = df_updated.drop(columns=["id","review_length","lastModeratedTime","submissionTime","userLocation","userNickname",'isVerifiedPurchaser','negativeFeedbackCount', 'lastModificationTime'])
    df_updated = df_updated.rename(columns={
    'text': 'comment_body',
    #'productName': 'post_title',
    #'productUrl': 'post_url',
    'positiveFeedbackCount': 'comment_likes'
    })




    df_updated['productName'] = df_updated['productName'].fillna('')
    return df_updated
```

```python

def correct_spelling_spellchecker(text):
    try:
        # Normalize curly apostrophes
        text = text.replace("’", "'").replace("‘", "'")

        tokens = word_tokenize(text)
        corrected_tokens = []

        for token in tokens:
            if token.isalpha():  # Only spell-check alphabetic words
                corrected = spell.correction(token)
                corrected_tokens.append(corrected if corrected else token)
            else:
                corrected_tokens.append(token)

        return rejoin_tokens(corrected_tokens)

    except Exception as e:
        print(f"Error processing text: {text[:100]}... | Skipping due to: {str(e)}")
        return text  # Return original text if error occurs

def rejoin_tokens(tokens):
    punctuation = {'.', ',', '!', '?', ':', ';', '(', ')', '"', "''", "``", "..."}
    clitics = {"n't", "'s", "'ve", "'re", "'m", "'d", "'ll"}

    result = ""
    for i, token in enumerate(tokens):
        if i == 0:
            result += token
        else:
            if token in punctuation or token in clitics:
                result += token
            else:
                result += " " + token
    return result
```

```python
# Clean text (skip post_title)
def clean_text(text):
    text = text.lower()
    text = re.sub(r'http\\S+', '', text)  # Remove URLs
    # Use raw string to retain spaces correctly
    #text = re.sub(r'[^a-zA-Z0-9.,!?$&@#\s]', '', text)  # Remove only unnecessary special characters
    text = re.sub(r"[^a-zA-Z0-9.,!?$&@#'\-\s]", '', text)
    text = re.sub(r'\s+', ' ', text).strip()  # Remove extra spaces
    return text

```

```python


# Tokenization function for text tokens
def tokenize_text(text):
    tokens = tokenizer.tokenize(text)  # Convert to subword tokens
    tokens = tokens[:max_length]  # Truncate if needed
    tokens += ['[PAD]'] * (max_length - len(tokens))  # Pad if needed
    return tokens

# Tokenization function for token IDs with padding
def tokenize_ids(text):
    # token_ids = tokenizer.encode(text, add_special_tokens=True)  # Convert to IDs with special tokens
    token_ids = tokenizer.encode(text, add_special_tokens=True, truncation=True, max_length=max_length)  # Truncate properly
    #token_ids = token_ids[:max_length]  # Truncate if needed
    token_ids += [tokenizer.pad_token_id] * (max_length - len(token_ids))  # Pad with [PAD] token ID
    return token_ids
```

```python
# NLP preprocessing
def NLPPreprocess_dataframe(dataframe):
    # Correct spelling
    # dataframe['comment_body'] = dataframe['comment_body'].apply(correct_spelling_spellchecker)



    # Ensure column has no NaN values
    dataframe['comment_body'] = dataframe['comment_body'].fillna("")

    # Correct spelling using parallel processing
    dataframe['comment_body'] = dataframe['comment_body'].swifter.apply(correct_spelling_spellchecker)

    # apply clean_text
    dataframe['comment_body'] = dataframe['comment_body'].apply(clean_text)

    #Remove stopwords
    stop_words = set(stopwords.words('english'))
    dataframe['comment_body'] = dataframe['comment_body'].apply(lambda x: ' '.join([word for word in x.split() if word not in stop_words]))

    # Lemmatization
    lemmatizer = WordNetLemmatizer()
    dataframe['comment_body'] = dataframe['comment_body'].apply(lambda x: ' '.join([lemmatizer.lemmatize(word) for word in x.split()]))

    # Initialize the tokenizer
    tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

    # Define max token length for padding/truncation
    max_length = 512


    # Apply tokenization to the DataFrame
    dataframe['tokens_text'] = dataframe['comment_body'].apply(tokenize_text)
    dataframe['tokens_ids'] = dataframe['comment_body'].apply(tokenize_ids)
    return dataframe
```
